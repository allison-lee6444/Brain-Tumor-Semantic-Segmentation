```python
# Function to create binary masks from COCO annotations and save them as label images
def create_and_save_mask(image_id, coco, height, width, save_path):
    mask = np.zeros((height, width), dtype=np.uint8)
    annotation_ids = coco.getAnnIds(imgIds=image_id)
    annotations = coco.loadAnns(annotation_ids)

    for annotation in annotations:
        for seg in annotation['segmentation']:
            poly = np.array(seg).reshape((int(len(seg) / 2), 2))
            rr, cc = polygon(poly[:, 1], poly[:, 0], (height, width))
            mask[rr, cc] = 1


    # Save mask to label directory
    mask_for_tif = (mask * 255).astype(np.uint8)
    Image.fromarray(mask_for_tif).save(save_path)  # Save the mask as a TIFF label image

def get_label_data(coco, data_dir, split, label_dir):
    img_ids = coco.getImgIds()
    for img_id in img_ids:
        img_info = coco.loadImgs(img_id)[0]
        img_path = os.path.join(data_dir, img_info['file_name'])
        # print(img_info['file_name'])
        label_path = os.path.join(label_dir, f"{split}_{os.path.splitext(img_info['file_name'])[0]}.tif")
        img = Image.open(img_path)
        create_and_save_mask(img_id, coco, img.size[1], img.size[0], label_path)

if is_process_data:
    get_label_data(coco_train, train_dir, 'train', train_label_dir)
    get_label_data(coco_val, val_dir, 'val', val_label_dir)
    get_label_data(coco_test, test_dir, 'test', test_label_dir)
```

```python
def convert_images_to_grey_tif(input_directory, file_extension="*.jpg", output_directory=None):
    image_paths = glob(os.path.join(input_directory, file_extension))

    # Convert and save each image
    for image_path in image_paths:
        img = Image.open(image_path)
        output_filename = os.path.basename(image_path).replace(".jpg", ".tif")
        output_path = os.path.join(output_directory, output_filename) if output_directory else image_path.replace(".jpg", ".tif")
        if img.mode != 'L':  # If the image is not already grayscale
            img = img.convert('L')
        img.save(output_path)

if is_process_data:
    # Convert images in train, val, and test directories to .tif format, and save in new dir
    convert_images_to_grey_tif(train_dir, "*.jpg", train_tif_dir)
    convert_images_to_grey_tif(val_dir, "*.jpg", val_tif_dir)
    convert_images_to_grey_tif(test_dir, "*.jpg", test_tif_dir)
```

```python
# Function to run automatic instance segmentation
def run_automatic_instance_segmentation(image, checkpoint_path, model_type="vit_b", device=None):
    """Automatic Instance Segmentation (AIS) by using a trained instance decoder in SAM."""
    predictor, segmenter = get_predictor_and_segmenter(
        model_type=model_type,
        checkpoint=checkpoint_path,
        device=device,
    )
    prediction = automatic_instance_segmentation(
        predictor=predictor,
        segmenter=segmenter,
        input_path=image,
        ndim=2,
    )
    return prediction

# Set the directory paths and checkpoint file path
# checkpoint_path = '/content/drive/MyDrive/CV/project/sam_hela/checkpoints/sam_hela/best.pt'
checkpoint_path = '/content/drive/MyDrive/CV/project/sam_hela_vit_h/checkpoints/sam_hela_vit_h/best.pt'

# Make sure the checkpoint exists
assert os.path.exists(checkpoint_path), "Please train the model first to run inference on the finetuned model."

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Function to calculate IoU between two masks
def calculate_iou(pred_mask, gt_mask):

    # Convert to binary mask (if needed)
    pred_mask = (pred_mask > 0.5).astype(np.uint8)

    # Ensure the ground truth mask is binary and 2D
    gt_mask = gt_mask.squeeze()  # Remove extra dimensions

    # Now calculate IoU
    intersection = np.logical_and(pred_mask, gt_mask).sum()
    union = np.logical_or(pred_mask, gt_mask).sum()
    return intersection / union if union > 0 else 0
```

```python
def most_likely_class(prediction):
  # Flatten the prediction array to make counting easier
  flat_prediction = prediction.flatten()

  # Use Counter to count the number of pixels for each class
  class_counts = Counter(flat_prediction)

  class_counts.pop(0, None)

  most_common_class = max(class_counts, key=class_counts.get)

  prediction = np.where(prediction == most_common_class, 1, 0)

  return prediction
```

```python
def add_box(prediction):
  from copy import deepcopy
  prediction = deepcopy(prediction)
  rows, cols = np.where(prediction == 1)

  if rows.size == 0 or cols.size == 0:
      return prediction

  # Calculate the bounding box
  min_row, max_row = np.min(rows), np.max(rows)
  min_col, max_col = np.min(cols), np.max(cols)
  prediction[min_row:max_row+1, min_col:max_col+1] = 1
  return prediction
```

```python
# Get a single image path in the test set
image_path = sorted(glob(os.path.join(test_tif_dir, "*.tif")))[0]
label_path = sorted(glob(os.path.join(test_label_dir, "test_*.tif")))[0]

# Run the segmentation on a single image
image = imageio.imread(image_path)
label = cv2.imread(label_path, cv2.IMREAD_GRAYSCALE)

# model_type = "vit_b"
model_type = "vit_h"

# Predicted instances
prediction = run_automatic_instance_segmentation(
    image=image, checkpoint_path=checkpoint_path, model_type=model_type, device=device
)

# Calculate the IoU
prediction = most_likely_class(prediction)
iou = calculate_iou(label, prediction)
print(f"IoU without box for the single test image: {iou:.4f}")

# Calculate the IoU of box prediction
prediction_box = add_box(prediction)
iou_box = calculate_iou(label, prediction_box)
print(f"IoU with box for the single test image: {iou_box:.4f}")
```

```python
# Mean Iou on test dataset
# Get all images in the test set
test_image_path = sorted(glob(os.path.join(test_tif_dir, "*.tif")))
test_label_path = sorted(glob(os.path.join(test_label_dir, "test_*.tif")))
iou_scores = []
iou_box_scores = []
i = 0

predictor, segmenter = get_predictor_and_segmenter(
        model_type=model_type,
        checkpoint=checkpoint_path,
        device=device,
    )


for image_path, label_path in zip(test_image_path, test_label_path):

    # Run the segmentation on a single image
    image = imageio.imread(image_path)
    label = cv2.imread(label_path, cv2.IMREAD_GRAYSCALE)

    if i%10==0:
      print(f"Now processing {i}/{len(test_image_path)} image")
    i+=1

    # Predicted instances
    prediction = automatic_instance_segmentation(
        predictor=predictor,
        segmenter=segmenter,
        input_path=image,
        ndim=2,
    )

    # Calculate the IoU
    prediction = most_likely_class(prediction)
    iou = calculate_iou(label, prediction)
    iou_scores.append(iou)

    prediction_box = add_box(prediction)
    iou_box = calculate_iou(label, prediction_box)
    iou_box_scores.append(iou_box)

mean_iou = sum(iou_scores) / len(iou_scores)
print(f"Mean IoU for the test set: {mean_iou:.4f}")

mean_iou_box = sum(iou_box_scores) / len(iou_box_scores)
print(f"Mean IoU for the box test set: {mean_iou_box:.4f}")
```
